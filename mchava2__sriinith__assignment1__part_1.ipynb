```python
#STEP - 2
#reading the csv file into a dataframe
penguins_df = pd.read_csv("/content/penguins.csv")
```

```python
#STEP - 2
#removing the rows containing the NA values
penguins_df.dropna(inplace=True)
penguins_df.reset_index(inplace=True)
penguins_df.drop(['index'],axis='columns',inplace=True)
penguins_df.isna().sum()
```

```python
#STEP - 3
#converting string datatype to categorical by using 'one - hot' encoding 
encoded_data = pd.get_dummies(penguins_df, columns = ['species','island'])
encoded_data
```

```python
#STEP - 4
#normalizing non-categorical features
#Finding min and max of each column
min_cols = encoded_data[['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g']].min()
max_cols = encoded_data[['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g']].max()
#rescaling
encoded_data[['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g']] = (encoded_data[['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g']] - min_cols) / (max_cols- min_cols)
```

```python
encoded_data
```

```python
#STEP- 5
#choosing the target variable and input varai
y = penguins_df[['sex']]
X = encoded_data[['bill_length_mm',
 'bill_depth_mm',
 'flipper_length_mm',
 'body_mass_g',
 'species_Adelie',
 'species_Chinstrap',
 'species_Gentoo',
 'island_Biscoe',
 'island_Dream',
 'island_Torgersen']]
```

```python
#STEP - 6
#creating data matrices(2d a) for X and Y
X_matx = X.to_numpy()
Y_matx = y.to_numpy()
```

```python
#STEP - 7
#dividing the dataset into training and testing

np.random.seed(10)
x=np.random.choice(a=np.arange(0,333,1),size=267,replace=False)
training_data = encoded_data.loc[x.tolist()]
testing_indexes = np.setdiff1d(penguins_df.index.values,x)
testing_data = encoded_data.loc[testing_indexes]
X_train,X_test = training_data.drop(labels=['sex'],axis='columns').to_numpy(),testing_data.drop(labels=['sex'],axis='columns').to_numpy()
y_train,y_test = training_data['sex'].to_numpy(),testing_data['sex'].to_numpy()



```

```python
# Encoding the Target column 'sex'
y_train = np.array(list(map(lambda sex: 0 if sex=="male" else 1,y_train.tolist())))
y_test = np.array(list(map(lambda sex: 0 if sex=="male" else 1,y_test.tolist())))
```

```python
#STEP - 8
#Printing the shape of X_train, y_train, X_test, y_test
print("Shape of X_train and X_test:",np.shape(X_train), np.shape(X_test))
print("Shape of y_train and Y_test:",np.shape(y_train),np.shape(y_test))
```

```python
#STEP - 9
class LogitRegression:
  def __init__(self,max_iterations,learning_rate):
    self.max_iterations = max_iterations
    self.learning_rate = learning_rate

  def sigmoid(self,z):
    self.weisum = z
    return 1 / (1 + np.exp(-self.weisum))

  def cost(self):
      loss=-self.y * np.log(self.pred_prob) - (1 - self.y) * np.log(1 - self.pred_prob)
      return loss.mean() 



  def gradient_descent(self):
    weight_change_factor = np.dot(self.X.T, (self.pred_prob - self.y)) / self.y.size
    return weight_change_factor

  def get_loss_values(self):
    return self.loss_values



  def fit(self,X_train,y_train):
    self.X = X_train
    self.y = y_train
    self.weights = np.random.uniform(0, 1, self.X.shape[1])
    self.loss_values=[]
    self.iter_labels=[]
    for i in range(self.max_iterations):
      self.z = np.dot(self.X, np.transpose(self.weights))
      self.pred_prob = self.sigmoid(self.z)
      self.weights = self.weights - (self.learning_rate*self.gradient_descent()) 
      self.loss_values.append(self.cost())

      self.itr_target_labels = list(map(lambda pred_prob: 1 if pred_prob>=0.5 else 0,self.z.tolist()))
      self.iter_labels.append(self.itr_target_labels)
      
    # checking for correctly labelled predictions
    self.correct_predictions = [np.array(pred_labels)==self.y for pred_labels in self.iter_labels]
    # training accuracy for all iterations
    self.acc = list(map(lambda correct_labels: sum(correct_labels)/len(self.y),self.correct_predictions))
     





  def predict(self, X_test):
    self.X_test = X_test
    self.pred_weisum = np.dot(self.X_test, np.transpose(self.weights))
    self.results = self.sigmoid(self.pred_weisum)
    ans = self.results>=0.5
    return ans
  

       
```

```python
# STEP - 10

# 10a.
# Creating Logistic Regressions models with different Learning rates and Iterations
model_1 = LogitRegression(max_iterations=10000,learning_rate=0.01)
model_2 = LogitRegression(max_iterations=10000,learning_rate=0.1)
model_3 = LogitRegression(max_iterations=10000,learning_rate=0.2)
model_4 = LogitRegression(max_iterations=20000,learning_rate=0.01)
model_5 = LogitRegression(max_iterations=20000,learning_rate=0.1)
model_6 = LogitRegression(max_iterations=20000,learning_rate=0.2)

# 10b.
# Training each model
model_1.fit(X_train,y_train)
model_2.fit(X_train,y_train)
model_3.fit(X_train,y_train)
model_4.fit(X_train,y_train)
model_5.fit(X_train,y_train)
model_6.fit(X_train,y_train)

# STEP - 11
# predictions for each model
model_1_predictions = model_1.predict(X_test)
model_2_predictions = model_2.predict(X_test)
model_3_predictions = model_3.predict(X_test)
model_4_predictions = model_4.predict(X_test)
model_5_predictions = model_5.predict(X_test)
model_6_predictions = model_6.predict(X_test)

# Accuracy for each model
model_1_acc = sum(model_1_predictions==y_test)/len(y_test)
model_2_acc = sum(model_2_predictions==y_test)/len(y_test)
model_3_acc = sum(model_3_predictions==y_test)/len(y_test)
model_4_acc = sum(model_4_predictions==y_test)/len(y_test)
model_5_acc = sum(model_5_predictions==y_test)/len(y_test)
model_6_acc = sum(model_6_predictions==y_test)/len(y_test)


print(f'Model 1 Test Accuracy:{round(model_1_acc*100,2)} Model 2 Test Accuracy:{round(model_2_acc*100,2)} Model 3 Test Accuracy:{round(model_3_acc*100,2)}\n Model 4 Test Accuracy: {round(model_4_acc*100,2)} Model 5 Test Accuracy{round(model_5_acc*100,2)} Model 6 Test Accuracy{round(model_6_acc*100,2)}')
```
